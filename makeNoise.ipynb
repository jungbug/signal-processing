```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

# 원본 WAV 파일 경로
original_file_path = "./output.wav"

# WAV 파일 읽기
sample_rate, audio_data = wavfile.read(original_file_path)

# 생성할 파일 개수
num_files = 5000

for i in range(num_files):
    # 노이즈 생성
    noise_amplitude = 0.1
    noise = np.random.normal(0, noise_amplitude, len(audio_data))

    # 오디오에 노이즈 추가
    noisy_audio = audio_data + noise

    # 새로운 WAV 파일 저장 경로
    output_file_path = f"./dataset/1/output_noisy_audio_{i}.wav"

    # 새로운 WAV 파일 생성
    wavfile.write(output_file_path, sample_rate, noisy_audio.astype(np.int16))


# 시간 배열 생성
time = np.arange(0, len(audio_data)) / sample_rate

# 오디오 파형 및 노이즈 플로팅
plt.figure(figsize=(12, 6))

# 원본 오디오
plt.subplot(2, 1, 1)
plt.plot(time, audio_data, label='Original Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

# 노이즈 추가된 오디오
plt.subplot(2, 1, 2)
plt.plot(time, noisy_audio, label='Noisy Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

# 노이즈 생성 함수
def generate_noise(duration, sample_rate, amplitude):
    length = int(duration * sample_rate)
    noise = amplitude * np.random.normal(size=length)
    return noise

# 노이즈 생성 설정
duration = 3.0  # 노이즈의 길이 (초)
sample_rate = 44100  # 샘플링 속도
amplitude = 2  # 노이즈의 진폭

# 생성할 파일 개수
num_files = 10

for i in range(num_files):
    # 노이즈 생성
    noise = generate_noise(duration, sample_rate, amplitude)

    # WAV 파일 저장 경로
    output_file_path = f"./generated_noise_{i}.wav"

    # 노이즈를 WAV 파일로 저장
    wavfile.write(output_file_path, sample_rate, noise.astype(np.int16))
```
