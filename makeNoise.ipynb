```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

#  WAV  
original_file_path = "./output.wav"

# WAV  
sample_rate, audio_data = wavfile.read(original_file_path)

#   
num_files = 100

for i in range(num_files):
    #  
    noise_amplitude = 0.1
    noise = np.random.normal(0, noise_amplitude, len(audio_data))

    #   
    noisy_audio = audio_data + noise

    #  WAV   
    output_file_path = f"./testset/1/output_noisy_audio_{i}.wav"

    #  WAV  
    wavfile.write(output_file_path, sample_rate, noisy_audio.astype(np.int16))


#   
time = np.arange(0, len(audio_data)) / sample_rate

#     
plt.figure(figsize=(12, 6))

#  
plt.subplot(2, 1, 1)
plt.plot(time, audio_data, label='Original Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

#   
plt.subplot(2, 1, 2)
plt.plot(time, noisy_audio, label='Noisy Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

#   
def generate_noise(duration, sample_rate, amplitude):
    length = int(duration * sample_rate)
    noise = amplitude * np.random.normal(size=length)
    return noise

#   
duration = 3.0  #   ()
sample_rate = 44100  #  
amplitude = 2  #  

#   
num_files = 100

for i in range(num_files):
    #  
    noise = generate_noise(duration, sample_rate, amplitude)

    # WAV   
    output_file_path = f"./testset/0/generated_noise_{i}.wav"

    #  WAV  
    wavfile.write(output_file_path, sample_rate, noise.astype(np.int16))
```
